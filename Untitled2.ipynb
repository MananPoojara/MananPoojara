```python
import pandas as pd
import numpy as np

# Upload CSV file
from google.colab import files
uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)
print("Original DataFrame:")
print(df)

# Dropping rows with any NaN values
df_dropped = df.dropna()
print("\nDataFrame after dropping rows with any NaN values:")
print(df_dropped)

# Filling NaN values with the mean of the column for numeric columns only
numeric_columns = df.select_dtypes(include=[np.number]).columns
df_mean = df.copy()
df_mean[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
print("\nDataFrame after filling NaN values with the mean:")
print(df_mean)


# Filling NaN values with the median of the column for numeric columns only
df_median = df.copy()
df_median[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())
print("\nDataFrame after filling NaN values with the median:")
print(df_median)


# Filling NaN values with the mode of the column for each column
df_mode = df.copy()
for column in df.columns:
    df_mode[column].fillna(df[column].mode()[0], inplace=True)
print("\nDataFrame after filling NaN values with the mode:")
print(df_mode)
```
